```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(x,y, color='Green', linewidth=5)
```

```python
# Printing Daimond structure
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(x,y, 'rD')
```

```python
plt.xlabel('Day')
plt.ylabel('Temprature')
plt.title('Weather')
plt.plot(x,y, marker='+')
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Weather")
plt.plot(days,max_t ,label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")
plt.legend(loc="best", shadow = True)
plt.grid()
```

# Understand of Bar chart

```python
import numpy as np
```

```python
company = ['Google','Amazon','MSFT','FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech stocks")
plt.bar(ypos-0.2,revenue, width=0.5 ,label = "revenue")
plt.bar(ypos+0.2,profit,width=0.5, label="Profit")
plt.legend()
```

# Histo Chart

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150] , rwidth=0.95, color="r", histtype="step")
```

```python
bloodsuger_men = [113,85,90,150,149,8,93,115,135,80,77,82,129]
bloodsuger_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel("Suger Label")
plt.ylabel("Total Number of Patients")
plt.title("Blood Sugar Analysis")
plt.hist([bloodsuger_men,bloodsuger_women],bins=[80,100,125,150] , rwidth=0.95, color=['Green','Orange'], label=['Men','Women'])
plt.legend()
```

#    Pie Chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home rent","Food","Phone/Internet Bills","Cars","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow='True', explode=[0,0.2,0,0,0], startangle=180)
```

# Scatter Chart

```python
x=[4,8,12,18,25,29]
y=[7,14,19,28,33,40]
plt.scatter(x,y, color="r", s=100)
plt.grid()
```

```python
x1=[4,8,12,18,25,29,30,35]
y1=[7,14,19,28,33,40,45,46]
x2=[2,4,12,15,17,20,30,35]
y2=[6,8,10,14,15,17,30,35]
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.scatter(x1,y1,color="Green", marker="+",label='X1Y1')
plt.scatter(x2,y2, color="Red", marker="*",label='X2Y2')
plt.grid()
plt.legend()
```

# Stack Chart/Area Chart

```python
week = [1,2,3,4,5]
shift1 = [40,35,26,23,50]
shift2 = [40,32,25,22,19]
shift3 = [40,30,22,20,18]
plt.stackplot(week, shift1, shift2, shift3, labels=['shift1', 'shift2', 'shift3'])
plt.legend()
```
